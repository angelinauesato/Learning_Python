```python
from faker.config import AVAILABLE_LOCALES
# AVAILABLE_LOCALES is a list of all supported locale strings
for locale in sorted(AVAILABLE_LOCALES):
    print(locale)
```

```python
# Filter locales that start with 'en_'
english_locales = [locale for locale in AVAILABLE_LOCALES if locale.startswith('en_')]
print(english_locales)
```

```python
fake = Faker(english_locales)

# Generate a complete user profile
def create_user_profile():
    return {
        # Personal details
        'name': fake.name(),
        'age': fake.random_int(min=18, max=80),
        'email': fake.email(),
        'phone': fake.phone_number(),
        
        # Location information
        'address': {
            'street': fake.street_address(),
            'city': fake.city(),
            'state': fake.state(),
            'zip': fake.zipcode(),
            'country': fake.country()
        },
        
        # Professional information
        'job': fake.job(),
        'company': fake.company(),
        
        # Additional details
        'username': fake.user_name(),
        'website': fake.url()
    }
```

```python
# Create and display a profile
profile = create_user_profile()
for _ in range(5):
    for key, value in profile.items():
        if isinstance(value, dict):
            print(f"\n{key.title()}:")
            for k, v in value.items():
                print(f"  {k.title()}: {v}")
        else:
            print(f"{key.title()}: {value}")
    print(" ###### ")
```
